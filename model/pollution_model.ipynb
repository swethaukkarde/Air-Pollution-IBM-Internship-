```python
import pandas as pd
import json
from sklearn.cluster import KMeans

# -----------------------------------
# LOAD DATASET
# -----------------------------------

df = pd.read_csv(r"C:\Users\Tharun Reddy\Desktop\AI_Air_pollution\city_day.csv")

# Keep only required columns
df = df[["City","Date","PM2.5","PM10","NO2","SO2","CO","O3","AQI"]]
df = df.dropna()

# Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# -----------------------------------
# 1) GET REAL LATEST AQI + GAS VALUES PER CITY
# -----------------------------------

latest = df.groupby("City").tail(1)

city_data = {}

for _, row in latest.iterrows():
    city = row["City"]
    city_data[city] = {
        "AQI": float(row["AQI"]),
        "PM25": float(row["PM2.5"]),
        "PM10": float(row["PM10"]),
        "NO2": float(row["NO2"]),
        "SO2": float(row["SO2"]),
        "CO": float(row["CO"]),
        "O3": float(row["O3"])
    }

# Save city AQI + gas values
with open("city_data.json", "w") as f:
    json.dump(city_data, f, indent=4)

print("city_data.json created successfully!")


# -----------------------------------
# 2) POLLUTION HOTSPOT CLUSTERING
# -----------------------------------

cluster_df = df.groupby("City")[["AQI","PM2.5","PM10"]].mean().reset_index()

# K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_df["Cluster"] = kmeans.fit_predict(cluster_df[["AQI","PM2.5","PM10"]])

# Convert clusters to dictionary
hotspots = dict(zip(cluster_df["City"], cluster_df["Cluster"]))

# Save hotspot clusters
with open("hotspots.json","w") as f:
    json.dump(hotspots, f, indent=4)

print("hotspots.json created successfully!")
```
